```python
''' get history price '''

## m1, m5, m15, m30, h1, h2, h6, h12, d1
interval = "d1"
asset_name = "bitcoin"

url = "http://api.coincap.io/v2/assets/"+asset_name+"/history?interval="+interval

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

# print(response.text)
```
